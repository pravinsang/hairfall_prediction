```python
x = data.drop(['Label'], axis = 'columns')
y = data['Label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data = scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy,mean_squared_error
```

```python
mlpb = MLPClassifier(max_iter= 10000, random_state=42,activation='relu',alpha=0.001,hidden_layer_sizes=(10,10),solver='adam')
mlpb.fit(x_train,y_train)


print("Number of nodes in input layer: ", mlpb.n_features_in_)

# Print the number of nodes in the output layer
print("Number of nodes in output layer: ", mlpb.n_outputs_)

# Print the number of hidden layers
print("Number of hidden layers: ", mlpb.n_layers_ - 2)

# Print the batch size
print("Batch size: ", mlpb.batch_size)

# Print the number of nodes in each hidden layer
print("Number of nodes in each layer: ", mlpb.hidden_layer_sizes)

# Print the maximum number of iterations
print("Maximum number of iterations: ", mlpb.max_iter)

# Print the activation function
print("Activation function: ", mlpb.activation)

# Print the optimizer
print("Optimizer: ", mlpb.solver)

# Predict the class probabilities for the test data
y_score = mlpb.predict_proba(x_test)[:, 1]

# Compute the false positive rate, true positive rate, and threshold values
fpr, tpr, thresholds = roc_curve(y_test, y_score)

# Compute the AUC
auc = roc_auc_score(y_test, y_score)

# Plot the ROC curve
plt.plot(fpr, tpr, label='MLP (AUC = {:.2f})'.format(auc))
plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
plt.savefig('roc.png')
```

```python
import tkinter as tk
from tkinter import ttk

# Define the input parameters
input_params_left = ["Gender", "Age", "Maritial Status", "Father's Hair Fall Problem", "Mother's Hair Fall Problem", "Grand Father's Hair Fall Problem", "Grand Mother's Hair Fall Problem", "Sibling Hair Fall Problem", "Pregnancy", "Radiation", "Hair Line Pattern"]
input_params_right = ["Hair Style", "Hair Density", "Average Hair Fall Rate", "Scalp Infection", "Itching", "Nutrition Intake", "Weight Loss", "Sleeping Pattern", "Chemical Products", "Medication", "Region"]


# Define the GUI window
root = tk.Tk()
root.title("MLP Classifier Predictor")
root.geometry("700x400")

# Define the input labels and entry fields for the left side
input_labels_left = []
input_entries_left = []
for i, param in enumerate(input_params_left):
    input_label = ttk.Label(root, text=param)
    input_label.grid(row=i, column=0, padx=10, pady=5, sticky=tk.W)
    input_labels_left.append(input_label)
    
    if param == "Gender":
        input_entry = ttk.Combobox(root, values=[0, 1, 2], state="readonly", width=18)
    elif param == "Maritial Status" or param == "Father's Hair Fall Problem" or param == "Pregnancy" or param == "Mother's Hair Fall Problem" or param == "Grand Father's Hair Fall Problem" or param == "Grand Mother's Hair Fall Problem" or param == "Sibling Hair Fall Problem" or param == "Radiation":
        input_entry = ttk.Combobox(root, values=[0, 1], state="readonly", width=18)
    else:
        input_entry = ttk.Entry(root, width=20)
        
    input_entry.grid(row=i, column=1, padx=10, pady=5)
    input_entries_left.append(input_entry)

# Define the input labels and entry fields for the right side
input_labels_right = []
input_entries_right = []
for i, param in enumerate(input_params_right):
    input_label = ttk.Label(root, text=param)
    input_label.grid(row=i, column=2, padx=10, pady=5, sticky=tk.W)
    input_labels_right.append(input_label)
    
    if param == "Hair Style" or param == "Scalp Infection" or param == "Itching" or param == "Weight Loss" or param == "Chemical Products" :
        input_entry = ttk.Combobox(root, values=[0, 1], state="readonly", width=18)
    elif param == "Hair Line Pattern" or param == "Hair Density" or param == "Nutrition Intake" or param == "Sleeping Pattern":
        input_entry = ttk.Combobox(root, values=[0, 1, 2], state="readonly", width=18)
    else:
        input_entry = ttk.Entry(root, width=20)
        
    input_entry.grid(row=i, column=3, padx=10, pady=5)
    input_entries_right.append(input_entry)
    
    
#img = tk.PhotoImage(file="cm.png")
#img_label = ttk.Label(root, image=img)
#img_label.grid(row=11, column=0, padx=10, pady=5, rowspan=2, sticky=tk.W)

# Define the prediction button
def predict():
    # Get the input values
    input_values_left = [int(entry.get()) if not isinstance(entry, ttk.Combobox) else int(entry.get()) for entry in input_entries_left]
    input_values_right = [int(entry.get()) if not isinstance(entry, ttk.Combobox) else int(entry.get()) for entry in input_entries_right]
    input_values = input_values_left + input_values_right
    
    print(input_values)
    
    input_values = np.array(input_values)
    
    input_values = input_values.reshape(1,22)
    input_values = scaler.transform(input_values)
    
    print(input_values)
    
    # Make the prediction
    prediction = mlp.predict(input_values)
    
    
    y_single_pred_proba = mlp.predict_proba(input_values)[0]
    # Show the prediction
    
    if prediction[0] == 0:
        output_label.config(text="Predicted Label = 0 Which means no Hair Loss Problem  With probability= {}".format(y_single_pred_proba[0]))
    else:
        output_label.config(text="Predicted Label = 1 Which means Hair Loss Problem With probability= {}".format(y_single_pred_proba[1]))

predict_button = ttk.Button(root, text="Predict", command=predict)
predict_button.grid(row=max(len(input_params_left), len(input_params_right))+1, column=1, padx=10, pady=10, columnspan=2)

# Define the output label
output_label = ttk.Label(root, text="")
output_label.grid(row=max(len(input_params_left), len(input_params_right))+2, column=1, padx=10, pady=10, columnspan=2)

# Run the GUI
root.mainloop()



```
