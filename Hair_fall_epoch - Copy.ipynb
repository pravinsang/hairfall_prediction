```python
x = data.drop(['Label'], axis = 'columns')
y = data['Label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data = scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter= 10000,activation='logistic',alpha=0.001,hidden_layer_sizes=(10,10),solver='adam')
mlp
```

```python
mlp.fit(x_train,y_train)
```

```python
pred = mlp.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

train_acc_list = [] 
test_acc_list = []
train_loss_list = []
test_loss_list = []

for epoch in range(20):
    
    mlp.fit(x_train,y_train)
    
    train_preds = mlp.predict(x_train)
    train_acc = accuracy_score(y_train, train_preds)
    train_acc_list.append(train_acc)
    train_loss = mean_squared_error(y_train, train_preds)
    train_loss_list.append(train_loss)

    test_preds = mlp.predict(x_test)
    test_acc = accuracy_score(y_test, test_preds)
    test_acc_list.append(test_acc)
    test_loss = mean_squared_error(y_test, test_preds)
    test_loss_list.append(test_loss)

```

```python
import matplotlib.pyplot as plt

plt.plot(train_acc_list, label='Train Accuracy')
plt.plot(test_acc_list, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy')
plt.legend()
plt.show()


```

```python
plt.plot(train_loss_list, label='train loss')
plt.plot(test_loss_list, label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Test Loss')
plt.legend()
plt.show()
```

```python
train_predictions = mlp.predict(x_train)
test_predictions = mlp.predict(x_test)
overall_train_acc = accuracy_score(y_train, train_predictions)
overall_test_acc = accuracy_score(y_test, test_predictions)
overall_train_mse = mean_squared_error(y_train, train_predictions)
overall_test_mse = mean_squared_error(y_test, test_predictions)

print("Overall Train Accuracy:", overall_train_acc)
print("Overall Test Accuracy:", overall_test_acc)
print("Overall Train MSE:", overall_train_mse)
print("Overall Test MSE:", overall_test_mse)
```
